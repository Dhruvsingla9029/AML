```python
from bing_image_downloader import downloader

# Download chicken images
downloader.download("chicken bird", limit=120, output_dir='dataset', adult_filter_off=True)

# Download duck images
downloader.download("duck bird", limit=120, output_dir='dataset', adult_filter_off=True)
```

```python
import os

os.rename('dataset/chicken bird', 'dataset/chicken')
os.rename('dataset/duck bird', 'dataset/duck')
```

```python
import os, shutil, random

def split_data(source, train, val, test):
    files = os.listdir(source)
    random.shuffle(files)

    train_files = files[:80]
    val_files = files[80:100]
    test_files = files[100:120]

    for f in train_files:
        shutil.copy(os.path.join(source, f), train)
    for f in val_files:
        shutil.copy(os.path.join(source, f), val)
    for f in test_files:
        shutil.copy(os.path.join(source, f), test)

# Create folders
for split in ['train', 'val', 'test']:
    for cls in ['chicken', 'duck']:
        os.makedirs(f'dataset/{split}/{cls}', exist_ok=True)

# Apply split
split_data('dataset/chicken', 'dataset/train/chicken', 'dataset/val/chicken', 'dataset/test/chicken')
split_data('dataset/duck', 'dataset/train/duck', 'dataset/val/duck', 'dataset/test/duck')
```
